```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#from matplotlib import interactive
#import PyQt5

%matplotlib notebook
```

```python
data = open('../Multiview_Action_3D_Dataset/dataset_full/Train/a09_s01_e01_skeleton_proj.txt')
```

```python
frames = {}

for line in data:
    frame = line.split()
    frame = np.array([float(x) for x in frame])
    if frame[0] in frames:
        frames[frame[0]] = np.append(frames[frame[0]], np.array(frame[1:]))
    else:
        frames[frame[0]] = np.array(frame[1:])


for key in frames:
    frames[key] = frames[key].reshape((20, 4))
```

```python
# each frame has data that looks like this
# joint num, x, y, z
frames[1]
```

```python
frame_num = 15
xs = frames[frame_num][:,1]
ys = frames[frame_num][:,2]
zs = frames[frame_num][:,3]
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
colors = ['blue' for i in range(20)]
colors[0] = 'red'
colors[7] = 'green'
colors[11] = 'green'
colors[15] = 'green'
colors[19] = 'green'
colors[3] = 'black'
ax.scatter(xs,ys,zs, color=colors)
for x, y, z, i in zip(xs, ys, zs, range(1,21)):
    ax.text(x,y,z,i)

```
