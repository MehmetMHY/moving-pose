```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import interactive
import PyQt5

%matplotlib notebook
```

```python
data = open('Multiview_Action_3D_Dataset/dataset_full/Train/a02_s04_e01_skeleton_proj.txt')
```

```python
frames = {}

for line in data:
    frame = line.split()
    frame = np.array([float(x) for x in frame])
    if frame[0] in frames:
        frames[frame[0]] = np.append(frames[frame[0]], np.array(frame[1:]))
    else:
        frames[frame[0]] = np.array(frame[1:])


for key in frames:
    frames[key] = frames[key].reshape((20, 4))
```

```python
# each frame has data that looks like this
# joint num, x, y, z
frames[1]
```

```python
frame_num = 20
xs = frames[frame_num][:,1]
ys = frames[frame_num][:,2]
zs = frames[frame_num][:,3]
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(xs[:],ys[:],zs[:])
```
